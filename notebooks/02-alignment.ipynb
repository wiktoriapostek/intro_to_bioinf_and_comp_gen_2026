## Exercise 1 — Implement Needleman–Wunsch and Smith–Waterman (Code)

**What you should do**

Implement two classic dynamic programming algorithms for sequence alignment:

* Needleman–Wunsch (NW) → global alignment
    * Aligns the entire sequence seq1 with the entire sequence seq2.
    * Best when sequences are expected to be similar across their full length.
* Smith–Waterman (SW) → local alignment
    * Finds the best matching subsequences inside seq1 and seq2.
    * Best when sequences may only share a conserved region (motif/domain).

### Needleman–Wunsch (Global Alignment)
#### Idea

NW tries to answer:

> “If I must align both sequences from start to end, what is the best alignment?”

It builds a dynamic programming matrix `S`, where `S[i, j]` stores the best possible score for aligning:

* `seq1[:i]` with `seq2[:j]`

#### Initialization matters

Because this is global alignment, gaps at the beginning count.
So the first row/column are filled with cumulative gap penalties.

Example intuition:

* Aligning first 3 chars of `seq1` with an empty sequence means 3 gaps → score = `3 * gap`

#### When to use NW

* Full-length homologous proteins
* Comparing similar genes/proteins end-to-end
* Cases where you care about the entire sequence, not just a motif

### Smith–Waterman (Local Alignment)
#### Idea

SW tries to answer:

> “What is the best matching region shared by these two sequences?”

Instead of forcing a full alignment, it finds the highest-scoring local region.
Each cell is computed as best of:
* diagonal, up, left and 0

Why 0?

If the alignment becomes bad (negative score), SW resets and starts a new local alignment region.
This is what makes it local.
