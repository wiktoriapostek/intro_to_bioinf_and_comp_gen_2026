### Task 2.3: Determination of amino acid enantiomers

Proteins in living organisms are almost exclusively built from **L-amino acids**.<br>
However, due to modeling errors, incorrect structure manipulation, or coordinate reflections, a protein structure may sometimes contain **D-amino acids**, which are mirror images of the natural form.

#### What does “L” or “D” mean?

Most amino acids are chiral molecules, meaning they exist in two mirror-image forms called **enantiomers**.
You can think of this like **left and right hands**:
- They look similar
- They are mirror images
- But they cannot be perfectly superimposed

In amino acids, chirality occurs at the **Cα (alpha carbon)**, which is bonded to four different groups:
- **N** - backbone nitrogen
- **Cα** – alpha carbon
- **C** – backbone carbonyl carbon
- **Cβ** – side chain carbon

These four atoms define the **local 3D geometry** of the residue.

If we construct a plane using **the atoms N–Cα–C**, we can determine whether **the Cβ atom lies on one side or the other of that plane**.
- If the orientation matches natural proteins → **L-amino acid**
- If it is flipped → **D-amino acid**

#### Why is this useful?

Checking chirality is a **simple quality control step** for protein structures.
If a structure contains many D-amino acids it may indicate:
- incorrect coordinate reflection
- modeling artifacts
- corrupted structures
 -incorrect atom ordering
